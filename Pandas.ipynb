```python
#modules
import numpy as np
import pandas as pd
```

### Concat operations

```python
#reading csv file
sample1 = pd.read_csv("sample_1.csv")
sample2 = pd.read_csv("sample_2.csv")
```

```python
sample1.head()
```

```python
sample2.head()
```

```python
#adding two data frames, default axis = 0 (veritical stacking)
pd.concat([sample1,sample2])
```

```python
#adding two data frames, default axis = 0 (veritical stacking)
pd.concat([sample1,sample2], axis=1)
```

### Apply 

```python
sample1.head()
```

```python
sample1["message_len"] = sample1.message.apply(len)
sample1.head()
```

```python
sample1["sample1_sqrt"] = sample1["c"].apply(np.sqrt)
```

```python
sample1.head()
```

```python
#apply with user def function

def max_min_dif(val):
    return val ** 1/4

sample1["app_cal"] = sample1["c"].apply(max_min_dif)
```
